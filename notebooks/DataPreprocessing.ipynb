```python
import pandas as pd
```

```python
CE = pd.read_csv('../../data/raw/CE.txt', delimiter='\t', header=None)
CP = pd.read_csv('../../data/raw/CP.txt', delimiter='\t', header=None)

EPS1 = pd.read_csv('../../data/raw/EPS1.txt', delimiter='\t', header=None)

FS1 = pd.read_csv('../../data/raw/FS1.txt', delimiter='\t', header=None)
FS2 = pd.read_csv('../../data/raw/FS2.txt', delimiter='\t', header=None)

PS1 = pd.read_csv('../../data/raw/PS1.txt', delimiter='\t', header=None)
PS2 = pd.read_csv('../../data/raw/PS2.txt', delimiter='\t', header=None)
PS3 = pd.read_csv('../../data/raw/PS3.txt', delimiter='\t', header=None)
PS4 = pd.read_csv('../../data/raw/PS4.txt', delimiter='\t', header=None)
PS5 = pd.read_csv('../../data/raw/PS5.txt', delimiter='\t', header=None)
PS6 = pd.read_csv('../../data/raw/PS6.txt', delimiter='\t', header=None)

SE = pd.read_csv('../../data/raw/SE.txt', delimiter='\t', header=None)

TS1 = pd.read_csv('../../data/raw/TS1.txt', delimiter='\t', header=None)
TS2 = pd.read_csv('../../data/raw/TS2.txt', delimiter='\t', header=None)
TS3 = pd.read_csv('../../data/raw/TS3.txt', delimiter='\t', header=None)
TS4 = pd.read_csv('../../data/raw/TS4.txt', delimiter='\t', header=None)

VS1 = pd.read_csv('../../data/raw/VS1.txt', delimiter='\t', header=None)
```

```python
labels = {0:'Cooler condition', 1:'Valve Condition', 2:'Internal pump leakage', 3:'Hydraulic accumulator / bar', 4:'stable flag'}
Y = pd.read_csv('../../data/raw/profile.txt', delimiter='\t', header=None).rename(columns=labels)
Y.index.name = 'Cycle ID'
```

downsample the high frequency sensors (lose a lot of data)

```python
def downsample(df, rate):
    return df.T[::rate].reset_index().T
```

```python
PS1_d, PS2_d, PS3_d, PS4_d, PS5_d, PS6_d, EPS1_d = [downsample(PS, 100) for PS in [PS1, PS2, PS3, PS4, PS5, PS6, EPS1]]
FS1_d, FS2_d = [downsample(FS, 10) for FS in [FS1, FS2]]
```

Merge downsampled Data on cycle index with name of Sensor
- multiindex the 

```python
def add_index(df, name):
    df = pd.concat([df.reset_index()], keys=[name], names=['Sensor', 'Cycle ID'])
    return df[1:]
```

```python
dfs = [PS1_d, PS2_d, PS3_d, PS4_d, PS5_d, PS6_d, EPS1_d, FS1_d, FS2_d, CE, CP, SE, TS1, TS2, TS3, TS4, VS1]
names = ['PS1_d', 'PS2_d', 'PS3_d', 'PS4_d', 'PS5_d', 'PS6_d', 'EPS1_d', 'FS1_d', 'FS2_d', 'CE', 'CP', 'SE', 'TS1', 'TS2', 'TS3', 'TS4', 'VS1']

PS1_d, PS2_d, PS3_d, PS4_d, PS5_d, PS6_d, EPS1_d, FS1_d, FS2_d, CE, CP, SE, TS1, TS2, TS3, TS4, VS1 = [add_index(x, y).drop(columns=['index']) for x, y in zip(dfs, names)]
```

```python
X = pd.concat(dfs, keys=names, names=['Sensor Name', 'Cycle ID'])[1::]
```

```python
full_df = X.join(Y, how='inner')
```

```python
mask, idx = full_df.index.get_loc_level(2004, level='Cycle ID')
full_df.loc[mask]
```

```python
full_df.to_csv('../../data/processed/downsampled/full_df.csv')
```

```python
full_df = full_df.swaplevel()
labels = ['Cooler condition', 'Valve Condition', 'Internal pump leakage',
           'Hydraulic accumulator / bar', 'stable flag']
X = full_df.drop(columns=labels)
y = full_df[labels]
X = X.sort_index().T.stack(level=['Cycle ID']).swaplevel()
full_df = pd.merge(X, y, left_on='Cycle ID', right_on='Cycle ID').drop_duplicates(keep='first')
X = full_df.drop(columns=labels)
y = full_df[labels]
full_df.to_csv('../../data/processed/downsampled/continuous_data.csv')
```
